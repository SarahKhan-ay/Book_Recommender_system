```python
books = books.drop(columns = ['id','best_book_id','work_id','books_count','isbn','isbn13','small_image_url','language_code','work_ratings_count','work_text_reviews_count','ratings_count','ratings_1',
       'ratings_2','ratings_3','ratings_4','ratings_5'])
```

```python
books.head()
```

```python
#Formatting
def collapse(str):
    str1 = ""
    for i in str:
        str1 = str1 + (i.replace(" ",""))
    return str1
```

```python
books['tags'] = books['authors'].apply(collapse) + ", " + books['original_title']
```

```python
books.head()
```

```python
books = books.rename(columns = {'original_publication_year': 'publication_year'})
books['publication_year'] = books['publication_year'].astype('int64')
```

```python
books.info()
```

```python
x = books['tags'].values.tolist()
y = ''
for i in x:
    y = y + i
len(set(y.split()))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english')
vector = cv.fit_transform(books['tags']).toarray()

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
def recommend(book):
    index = books[books['title'] == book].index[0]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    recommended_book_names = []
    image_list = []
    author_list = []
    publication_year = []
    rating = []
    for i in distances[1:6]:
        author_list.append(books.iloc[i[0]].authors)
        publication_year.append(books.iloc[i[0]].publication_year)
        rating.append(books.iloc[i[0]].average_rating)
        image_list.append(books.iloc[i[0]].image_url)
        recommended_book_names.append(books.iloc[i[0]].title)
    return recommended_book_names, author_list, publication_year, rating, image_list   
      
```

```python
recommend('The Hunger Games (The Hunger Games, #1)')
```
